### 1. Calcule los errores absolutos y relativo en las aproximaciones de p por p*

* Error absoluto:  $E_{abs} = |p-p^*|$
* Error relativo:  $E_{rel} = \frac{|p-p^*|}{|p|}$
  
  Donde:
  * p -> valor exacto
  * p* -> valor aproximado

```python
import math

def error_absoluto(p, p_aprox):
    return abs(p - p_aprox)

def error_relativo(p, p_aprox):
    return abs(p - p_aprox)/abs(p)

print("a) p = œÄ, p* = 22/7 ")
p = math.pi
p_aprox = 22/7
print("Error absoluto:", error_absoluto(p, p_aprox))
print("Error relativo:", error_relativo(p, p_aprox))
print()

print("b) p = œÄ, p* = 3.1416 ")
p = math.pi
p_aprox = 3.1416
print("Error absoluto:", error_absoluto(p, p_aprox))
print("Error relativo:", error_relativo(p, p_aprox))
print()

print("c) p = e, p* = 2.718 ")
p = math.e
p_aprox = 2.718
print("Error absoluto:", error_absoluto(p, p_aprox))
print("Error relativo:", error_relativo(p, p_aprox))
print()

print("d) p = ‚àö2, p* = 1.414 ")
p = math.sqrt(2)
p_aprox = 1.414
print("Error absoluto:", error_absoluto(p, p_aprox))
print("Error relativo:", error_relativo(p, p_aprox))
```
